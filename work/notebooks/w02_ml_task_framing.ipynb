```python
# The target variable we want to predict is 'is_spam'.
# This is a binary label: 1 if the message is spam, 0 if it is not spam.
# In a real-world scenario, this label would come from human annotations (observed outcome)
# or a highly accurate, pre-existing spam filter (defined rule).

target_variable = 'is_spam'

print(f"Target variable: {target_variable}")
print(f"Description: Binary classification (0: not spam, 1: spam)")
print(f"Source of label (for our dummy data): Defined within the dataset, representing ground truth.")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# For spam classification, we want to minimize false positives (legitimate emails marked as spam).
# Therefore, a high precision is often desired.
# Recall is also important to catch as much spam as possible.
# A balanced metric like F1-score combines both.

# Primary metric: Precision
# We want a high precision (e.g., > 0.95 or 95%) to ensure that when we flag an email as spam, it genuinely is spam.
# This avoids legitimate emails being lost in the spam folder.

# Secondary metrics:
# Recall: Also important to ensure we're catching a significant portion of actual spam.
# F1-score: A harmonic mean of precision and recall, useful for imbalanced datasets and providing a single metric for overall performance.
# Accuracy: While intuitive, it can be misleading in imbalanced datasets (e.g., if only 1% of emails are spam, classifying everything as 'not spam' yields 99% accuracy).

print("Primary Success Metric: Precision")
print("Target: > 0.95 (meaning less than 5% of flagged spam emails are actually legitimate)")
print("Secondary Metrics: Recall, F1-score")
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

# Create a dummy DataFrame representing the unit of analysis
data = {
    'message_id': [1, 2, 3, 4, 5],
    'text_content': [
        'Hello, how are you?',
        'Win a free iPhone now! Click here.',
        'Meeting reminder for tomorrow.',
        'Your account has been compromised. Verify here.',
        'Thanks for the update.'
    ],
    'text_length': [19, 35, 29, 49, 25],
    'num_links': [0, 1, 0, 1, 0],
    'has_spam_keywords': [False, True, False, True, False],
    'is_spam': [0, 1, 0, 1, 0] # 0 for not spam, 1 for spam
}

df_analysis = pd.DataFrame(data)

print("Unit of Analysis DataFrame:")
print(df_analysis.head())
print(f"\nShape of the DataFrame: {df_analysis.shape}")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Let's try to define a simple rule-based spam detector
def rule_based_spam_detector(row):
    # Rule 1: If there's a link, it might be spam
    if row['num_links'] > 0:
        return 1
    # Rule 2: If it has known spam keywords, it's spam
    if row['has_spam_keywords']:
        return 1
    # Rule 3: Very short or very long messages could be outliers, but hard to define a fixed threshold reliably
    # For simplicity, let's say if text_length is very small and no other indicators, it's not spam
    # This is where rules become brittle: a short message like 'Claim your prize!' is spam, but 'Hi!' is not.

    return 0 # Default to not spam if no rules are triggered

# Apply the rule-based detector to our dataframe
df_analysis['rule_based_prediction'] = df_analysis.apply(rule_based_spam_detector, axis=1)

print("Comparing Rule-Based Predictions with Actual Labels:")
print(df_analysis[['text_content', 'is_spam', 'rule_based_prediction']])

# Highlight cases where the rule-based system fails (e.g., false positives or false negatives)
# For a real dataset, this would show more discrepancies. Here, our dummy data aligns well for these simple rules.
# In a real scenario, a message like 'Your package is delayed: [link]' might be legitimate but flagged by Rule 1.
# Or 'Amazing offer, call now!' might be spam but missed by these rules if no keywords/links are explicitly coded.

print("\nWhy rules are messy:")
print("1. The definition of 'spam keywords' is constantly evolving and hard to maintain manually.")
print("2. Legitimate messages can contain links (e.g., password reset, order confirmation).")
print("3. Spammers actively try to evade simple rules by changing phrasing or obscuring links.")
print("4. Context is often crucial (e.g., a message from a bank vs. a random sender).")
print("Machine Learning can learn these complex, non-linear patterns and adapt to new spam tactics more effectively than a fixed set of 'if-then' rules.")
```
